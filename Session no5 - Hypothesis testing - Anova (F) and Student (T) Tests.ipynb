# ANOVA: Exercise 1

Researchers want to test a new anti-anxiety medication. They split participants into three conditions (0mg, 50mg and 100mg), then ask them to rate their anxiety level on a scale from 1 to 10. Are there any differences between the three conditions using alpha = 0.05.

(Source: [YouTube Course](https://www.youtube.com/watch?v=51QZa7b0Ozk))


### Answer:

So, from what I'm understanding here, this test is about comparing more than 2 groups. The null hypothesis is to say that the mean of the 3 groups in this exercise are equals. More specifically, we're saying that the values of the 3 samples don't differ enough to be able to assert that the medication actually have an effect on the condition to cure...

***With that being said:***
h_1 is saying that these means are not equals. So it's a little bit like the previous Z test we done when comparing two averages the main difference here, is that you're comparing more than just 2 groups...

As usual, we set alpha to 5%

```python
# First step: create the dataset:
df = pd.DataFrame({'0mg':[9,8,7,8,8,9,8], '50mg':[7,6,6,7,8,7,6], '100mg':[4,3,2,3,4,3,2]})
df
```

Quick hint: We're looking for the F statitstic (last field of the tab). We'll find it using the tab's other fields which formulas can be found below. The plan here is that, once we get the F statistic, we compare it to a "critical value" (I'll define what this is shortltly). If F is greater than the critical value, then we can reject the null hypothesis that states that the variations we see in our dataset come from the randomness of the experiment...

The critical value is calculated using alpha (like in every previous hypothses tests we've ran before). Here, we'll be using a table to lookup for that critical value. Check what is happening below for more detail:

```python
# The first step of the ANOVA (once we got all the parameters from above) is to calculate the degrees of freedom
    # Degrees of freedom between = a - 1
    # Degrees of freedom within = N - a
    # Degrees of freedom total = N - 1
    # with "N" (capital N) being the total number of values in your dataset, "n" (small "n") being the number of values per columns and a being the number of columns/variables in your dataset...

# So in our example:
N = df.shape[0]*df.shape[1]
n = df.shape[0]
a = df.shape[1]

# Let's just write small functions to calculate those values:

def df_between(df):
    res = df.shape[1] - 1
    return res

def df_within(df):
    res = df.shape[0]*df.shape[1] - df.shape[1]
    return res

def df_total(df):
    res = df.shape[0]*df.shape[1] - 1
    return res
    
    

df_between = df_between(df)
df_within = df_within(df)
df_total = df_total(df)

# Values to print
df_between
df_within
df_total
    
```

One thing you'll notice and that's always true (if you find different results, you probably got your calculus wrong...):
- dfbetween+dfwithin = dftotal

***Now the next step is to "state the decision rule"***
- basically we need to use df_between and df_within to lookup the critical value. Like I said above, the statistics we'll calculate will have to be greater than that critical value to reject H_0     
- here dfbetween = 2 and dfwithin = 18 like we've seen above... we'll just look those values up in the F table.. (2:44 in the yt vid...)     
- So with our "critical value" being 3.5546, we can say that if F (the stat we'll calculate below) is greater than 3.5546, we'll be able to reject the null hypothesis

```python
# Let's start filling out our tab with the values already calculated above:
df2 = pd.DataFrame({'SS':[" "," "," "], 'df':[" "," "," "], 'MS':[" "," "," "], 'F':[" "," "," "]}, index = ["Between", "Within", "Total"])
df2.df = [df_between,df_within,df_total]
df2
```

```python
# Now the whole point is to calculate the remaining ones in order to finally get to the F statistic we're looking for
# Knowing the equations for the SS (Sum of Squares) Between/Within/Total (maybe find a picture or something...) I just wrote the following 3 functions:


def get_SS_between(df):
    N = 0
    col_sums = []
    sum_all_values_in_df =0
    sum_all_values_squared_in_df = 0
    n = len(df)
    for i in range(len(df.columns)):
        col = df.iloc[:, i]
        col_sum = col.sum()
        N = N+len(col)
        col_sums.append(col_sum)
    for i in range(len(col_sums)):
        sum_all_values_in_df =col_sums[i]+ sum_all_values_in_df
        sum_all_values_squared_in_df = np.square(col_sums[i])+ sum_all_values_squared_in_df
    SS_between = (sum_all_values_squared_in_df/n) - np.square(sum_all_values_in_df)/N
    return SS_between
    
SS_between = get_SS_between(df)
SS_between = round(SS_between, 2)

def get_SS_within(df):
    N = 0
    col_sums = []
    sum_all_values_squared_in_df = 0
    squared_individual_values = np.square(df.values)
    tot_squared_individual_values = 0
    n = len(df)
    for i in range(len(df.columns)):
        col = df.iloc[:, i]
        col_sum = col.sum()
        N = N+len(col)
        col_sums.append(col_sum)
    for i in range(len(col_sums)):
        sum_all_values_squared_in_df = np.square(col_sums[i])+ sum_all_values_squared_in_df
    res_temp = sum_all_values_squared_in_df/n
    for i in squared_individual_values:
        tot_squared_individual_values = i + tot_squared_individual_values
    SS_Within = tot_squared_individual_values.sum() - res_temp
    return SS_Within

SS_within = get_SS_within(df)
SS_within = round(SS_within, 2)

def get_SS_total(df):
    N = 0
    col_sums = []
    sum_all_values_in_df =0
    squared_individual_values = np.square(df.values)
    tot_squared_individual_values = 0
    n = len(df)
    for i in range(len(df.columns)):
        col = df.iloc[:, i]
        col_sum = col.sum()
        N = N+len(col)
        col_sums.append(col_sum)
    for i in range(len(col_sums)):
        sum_all_values_in_df =col_sums[i]+ sum_all_values_in_df
    res_temp = np.square(sum_all_values_in_df)/N
    for i in squared_individual_values:
        tot_squared_individual_values = i + tot_squared_individual_values
    SS_total = tot_squared_individual_values.sum() - res_temp
    return SS_total

SS_total = get_SS_total(df)
SS_total = round(SS_total, 2)

# Values to print out...
SS_between
SS_within
SS_total

# Alright, let's just fill up those values into our tab from above:
df2.SS = [SS_between,SS_within,SS_total]
df2
```

```python
# To get the values of the remaining column: "M of Squares", we just need to divide the Sum of Square values by the corresponding degrees of freedom:
# so we'd just get here:
df2.MS = df2.SS / df2.df
df2.MS = df2.MS.round(2)
df2
```

```python
# We usually don't fill out the MS total value... so let me just remove it..
df2.loc["Total", "MS"] = ""
df2
```

```python
# We're finally reaching the end!! The only thing left is to get our F value! Don't pay attention to the fact that there also are 3 rows in F, we're just going to fill the first one...
# F is equal to MS Between/MS Within. So we'd have:

df2.loc["Between", "F"] = round(df2.loc["Between", "MS"] / df2.loc["Within", "MS"], 2)
df2

```

### Conclusion
Now that we have our F statistic, the idea is to compare it to the critical value using the right distribution.
we can just conclude by saying that our F value is far beyond our critical value, so we can easily reject H_0 and say that the 3 means are not equal.

### A cleaner way of concluding would be: 
- "the three conditions differed significantly on anxiety level 
- "Mathematically" we'd have: F(2,18) = 86.56, p<0.05

### Disclaimer: 
Watch out, this test only allows us to say that these 3 groups are different "somehow", we can't tell what actually makes them different. To figure that out, we'd have to perform a "post hoc" test..
